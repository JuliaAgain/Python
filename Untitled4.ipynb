3. Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
import numpy as np
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
import seaborn as sns
sns.heatmap(corr_matrix, annot=True, linewidths = 0.5, cmap = 'coolwarm');
```

Стоимость недвижимости зависит от жилой площади,оценки недвижимости (grade), количества ванных, жилой площади у 15 соседей, количества квдратных метров над землей.


```python
data = df.groupby('grade').agg({'price' : 'mean'}).reset_index()
data
```

```python
plt.figure(figsize = (8, 6))
plt.plot(data['grade'], data['price'])
plt.xlabel('Grade')
plt.ylabel('Цена')
plt.grid();
```

Средняя цена недвижимости повышается при повышении оценки (Grade) недвижимости.

```python
data1 = df.groupby('sqft_living').agg({'price' : 'mean'}).reset_index()
data1
```

```python
plt.figure(figsize = (8, 6))
plt.plot(data1['sqft_living'], data1['price'])
plt.xlabel('Жилая площадь')
plt.ylabel('Цена')
plt.grid();
```

В целом в диапазоне значений жилой площади от 500 до 8000 отмечается рост средней стоимости недвижимости с ростом жилой площади.
